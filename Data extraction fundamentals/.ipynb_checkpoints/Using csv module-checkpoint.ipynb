```python
import os
import pprint
import csv

DATADIR = ""
DATAFILE = "beatles-diskography.csv"
```

```python
def parse_csv(datafile):
    data = []
    n = 0
    with open(datafile, 'r') as sd:
        r = csv.DictReader(sd)
        print(r)
        for line in r:
            data.append(line)
    return data
```

```python
datafile = os.path.join(DATADIR, DATAFILE)
data = parse_csv(datafile)
data
```
